```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#Load the image
lena_img = cv2.imread('lena_image.jpg', cv2.IMREAD_GRAYSCALE)

plt.imshow(lena_img, cmap='gray')
plt.title('Lena Image')
```

```python
def no_padding_conv(img, filter):
    Ih, Iw = img.shape
    Fh, Fw = filter.shape
    new_img = np.zeros((Ih - Fh + 1, Iw - Fw + 1), dtype=np.float32)

    for i in range(Fh // 2, Ih - Fh // 2):
        for j in range(Fw // 2, Iw - Fw // 2):
            window = img[i - Fh // 2 : i + Fh // 2 + 1, j - Fw // 2 : j + Fw // 2 + 1]
            new_img[i - Fh // 2, j - Fw // 2] = np.sum(window * filter)

    return np.clip(new_img, 0, 255).astype(np.uint8)

def prewitt_edge_detect(img):
    Px = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]], dtype=np.float32)
    Py = np.array([[1, 0, -1], [1, 0, -1], [1, 0, -1]], dtype=np.float32)
    thresh = 20

    img = img.astype(np.float32)
    fx = no_padding_conv(img, Px).astype(np.float32)
    fy = no_padding_conv(img, Py).astype(np.float32)

    edge_img = np.sqrt(fx ** 2 + fy ** 2)
    edge_img = np.clip(edge_img, 0, 255).astype(np.uint8)

    edge_img = cv2.resize(edge_img, (img.shape[1], img.shape[0]), interpolation=cv2.INTER_NEAREST)

    return edge_img

def sharpen(img, edge_img, constant=1.5):
    result = img.astype(np.float32) + constant * edge_img.astype(np.float32)
    return np.clip(result, 0, 255).astype(np.uint8)
```

```python
if __name__ == "__main__":
    img = cv2.imread("lena_image.jpg", cv2.IMREAD_GRAYSCALE)

    edge_img = prewitt_edge_detect(img)

    sharpened_img = sharpen(img, edge_img, constant=1.5)

    plt.subplot(131)
    plt.imshow(img, cmap='gray')

    plt.subplot(132)
    plt.imshow(edge_img, cmap='gray')

    plt.subplot(133)
    plt.imshow(sharpened_img, cmap='gray')
```
